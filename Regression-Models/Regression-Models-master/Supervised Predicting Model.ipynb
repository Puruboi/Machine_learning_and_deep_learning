```
# Improting Libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras
print('Libraries imported')
```

```
# Defining a predicting function with linear regression 
# and training the model with study hours as input and scores as output training it with few trails to predict score for future unknown values
def score_model(y_new):
    xs = np.array([2.5,5.1,3.2,8.5,3.5,1.5,9.2,5.5,8.3,2.7,7.7,5.9,4.5,3.3,1.1,8.9,2.5,1.9,6.1,7.4,2.7,4.8,3.8,6.9,7.8],dtype=float)
    ys = np.array([21,47,27,75,30,20,88,60,81,25,85,62,41,42,17,95,30,24,67,69,30,54,35,76,86],dtype=float)
    model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer='sgd',loss=tf.keras.losses.MeanAbsoluteError())
    model.fit(xs, ys, epochs=700)
    return model.predict(y_new)
```

```
# Calling the function to predict score for custom input 
n = float(input("Enter the study hour for which the score to be predicted : "))
prediction = score_model([n])
print("The predicted score with respect to the study hours  is : {} ".format(prediction) )
```
